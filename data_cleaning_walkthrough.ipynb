```python
import re # extract latitude from column
def ext (col):
    col = re.findall("\(.+\)", col)
    col = col[0].split(',')[0].replace('(', '')
    return col
```

```python
data['hs_directory']['lat'] = data['hs_directory']['Location 1'].apply(ext)
```

```python
import re # extract latitude from column
def ext_lon (col):
    col = re.findall("\(.+\)", col)
    col = col[0].split(',')[1].replace(')', '').strip()
    return col
```

```python
data['hs_directory']['lon'] = data['hs_directory']['Location 1'].apply(ext_lon)
```
